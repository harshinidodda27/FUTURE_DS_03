```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

print("Libraries loaded successfully!")
```

```python
df = pd.read_csv('bank[1].csv')
print(df.head())
print(df.columns)
```

```python
print("Shape:",df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['deposit'].value_counts()
```

```python
conversion_rate = (df['deposit'] == 'yes').mean() * 100
print(f"Conversion Rate: {conversion_rate:.2f}%")
```

```python
sns.countplot(x='deposit', data=df)
plt.title('Customer Conversion Distribution')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='job', hue='deposit', data=df)
plt.xticks(rotation=45)
plt.title('Conversion by Job')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='contact', hue='deposit', data=df)
plt.title('Conversion by Contact Method')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='education', hue='deposit', data=df)
plt.title('Conversion by Education Level')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='poutcome', hue='deposit', data=df)
plt.title('Conversion by Previous Campaign Outcome')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['campaign'], bins=20)
plt.title('Campaign Contact Distribution')
plt.xlabel('Number of Contacts')
plt.ylabel('Number of Customers')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['age'], bins=20)
plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.show()
```

# Key Insights

- Dataset contains 11,162 customer records.
- No missing values were found.
- Overall conversion rate is 47.38%.
- More customers declined the deposit offer than accepted it.
- Conversion varies across job categories.
- Contact method affects conversion performance.
- Previous successful campaign outcomes lead to higher conversion rates.
- Certain age groups contribute more to conversions.

# Recommendations

- Focus on customer groups with higher conversion rates.
- Use the most effective contact methods.
- Re-target customers who responded positively before.
- Reduce excessive campaign contacts.
- Personalize marketing campaigns based on customer profiles.
